# Gráficos

```python
from matplotlib import pyplot as plt
Auto['mpg'].plot(kind='hist', bins=20, title='mpg')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
Auto.plot(kind='scatter', x='displacement', y='horsepower', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
